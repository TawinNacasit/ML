```python
from pathlib import Path

data = {"path": [], "class":[]}

for class_ in Path("cats_vs_dogs").iterdir():
    paths = class_.glob("*")#("*.jpg")
    for path in paths:
        data["path"].append(path)
        data["class"].append(class_.name)

cats_vs_dogs = pd.DataFrame.from_dict(data)
cats_vs_dogs
#list(Path("cats_vs_dogs").iterdir())
```
