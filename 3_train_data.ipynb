# Weak Supervision
- กำหนดเงื่อนไขง่ายๆ แต่หลายฟังก์ชั่น หลายประเด็น หลายคอลลัม
 

```python
#complaints["ประเด็นที่ร้องเรียน"] #ถ้าไมมีเฉลยให้สังเกตุคำ เช่น น้ำท่วมเป็นสาธารณะภัย เพื่อแปะเฉลย
complaints = pd.read_excel("complaints.xlsx")

def label_function(x):
    if "น้ำ" in x["ประเด็นที่ร้องเรียน"]:
        return "ภัย"
    elif "ถนน" in x["ประเด็นที่ร้องเรียน"]:
        return "ซ่อม"

#เช็คหมวดหมู่
#complaints.columns

complaints.dropna(subset=["ประเด็นที่ร้องเรียน"]).apply(label_function, axis=1) 
```

- Undersampling

```python
from imblearn.under_sampling import RandomUnderSampler

house_prices = pd.read_csv("house_prices.csv", index_col=)

complaints = pd.read_excel("complaints.xlsx")
complaints = complaints.dropna()

x, y = train_test_split(
    complaints,
    test_size=0.3,
    stratify=complaints["หมวดหมู่ปัญหา (เลือก)"],
    random_state=1234,
)

resampler = RandomUnderSampler()
resampler.fit_resample(x, y)
```
