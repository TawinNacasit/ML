# Feature Engineering

```python
import pandas as pd

house_prices = pd.read_csv("house_prices.csv")
#house_prices.isna() ##เช็คค่าว่าง
#house_prices.isna().any() ##เช็คว่าที่เรียกมาอันไหนเป็นว่างหรือไม่ว่าง 
#house_prices.isna().any(axis=1) ##เช็คแนวคอลัม .any(axis=1) เช็คแนวแถว

# selector = house_prices.isna().any(axis=1)
# house_prices.isna().any(axis=1)[selector]
house_prices.dropna("Unnamed: 0", axis=1)
house_prices
```

# ข้อมูลหายแก้อย่างไร

- Dropna ลบข้อมูลออก

```python
house_prices.dropna(axis=0, subset=["location", "city"], how= "any") #inplace=True แก้ไขทับตัวแปรเดิม
```

- Fillna ใส่ค่าเพิ่ม

```python
imputation = house_prices["city"].fillna("Bangkok") #ในคอลัม city ถ้าว่าง ให้ใส่ Bangkok
imputation
```

# Scaling

```python
house_prices.describe()  # ดูสถิติข้อมูล
```

```python
#!pip install scikit-learn
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(house_prices[["Area_in_Marla"]]) #สั่งให้เรียนรู้ เรียกหลังสร้าง Object ในกรณีนี้อยากให้เรียนรู้ข้อมูลพื้นที่(คอลลัม)

house_prices["area_scaled"] = scaler.transform(house_prices[["Area_in_Marla"]]) #ข้อมูลพื้นที่ ที่ลด scal ลงให้อยู่ในช่วง 0-1

```

```python
from sklearn.preprocessing import StandardScaler

zmv = StandardScaler()
zmv.fit(house_prices[["bedrooms"]])

house_prices["bedrooms_scaled"] = zmv.transform(house_prices[["bedrooms"]])
```

- Log transformation

```python
#house_prices["price"].plot(kind="hist")
#house_prices["price"].describe()

import numpy as np

house_prices["price_log"] = np.log10(house_prices["price"])
house_prices
```

# Discretization
 - แบ่งกลุ่มข้อมูล

    - ตัวเลข
    
    - ข้อความ
    

## 1. ตัวเลข

```python
from sklearn.preprocessing import KBinsDiscretizer #แบ่งเป็นถัง

discretizer = KBinsDiscretizer(n_bins= 5 , encode="ordinal") #n_bins แบ่งเป็นกี่ช่วง
discretizer.fit(house_prices[["price_log"]])
house_prices["price_log_bit"] = discretizer.transform(house_prices[["price_log"]]) # ผิด มี5คอลลัม ยัดเข้า 1 คอลลัมไม่ได้ ต้องเพ่ิ่ม encode = "ordinal"
```

## 2. ข้อความ

### One-hot

```python
house_prices_oh = pd.get_dummies(house_prices, columns=["purpose"], dtype=int) 
```

# Feature crossing

```python
house_prices_oh["bath_bedrooms"] = house_prices_oh["baths"] * house_prices_oh["bedrooms"]
```

```python
house_prices_small = house_prices_oh[["price", "baths", "bedrooms", "Area_in_Marla"]]

house_prices_small
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(house_prices_small, test_size=0.3, random_state=1234)
len(train), len(test)
```

# แบ่งข้อมูลก่อน train

```python
#from sklearn.model_selection import train_test_split

random_number = np.random.random(size=house_prices_small.shape[0])
house_prices_small["random"] = random_number

train, test = train_test_split(house_prices_small, test_size=0.3, random_state=1234)
len(train), len(test)
```

```python
from sklearn.inspection import permutation_importance
from sklearn.linear_model import LinearRegression

features = ["baths", "bedrooms", "Area_in_Marla", "random"]
price = "price"

X_train, Y_train = train[features], train[price]

model = LinearRegression()
model.fit(train[features], train[price])

importances = permutation_importance(model, test[features], test[price])
importances = pd.Series(importances.importances_mean, index=features)
importances.sort_values(ascending=False)
importances
```
